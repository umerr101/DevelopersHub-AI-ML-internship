# Task 1: Exploring and Visualizing a Simple Dataset 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
```

```python
# Scatter-Plot

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data = data,
    x = 'sepal_length',
    y = 'sepal_width',
    hue = 'species',
    palette = 'viridis'
)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
#Histogram

data.hist(figsize=(10, 8), bins=15)
plt.suptitle("Feature Distributions")
plt.tight_layout()
plt.show()
```

```python
# Box Plot

plt.figure(figsize=(10, 6))
sns.boxplot(data=data.drop('species', axis=1))
plt.title('Box Plots of Iris Features')
plt.show()
```

## Key Findings from Iris Dataset Visualization

### Dataset Overview
- **Total Samples**: 150 flowers across 3 species (Setosa, Versicolor, Virginica)
- **Features Measured**: Sepal Length, Sepal Width, Petal Length, Petal Width

### Findings from Scatter Plot (Sepal Length vs Sepal Width)
- **Setosa flowers** (blue) cluster tightly with shorter sepal lengths (4.3-5.8 cm) and wider sepals (2.3-4.4 cm)
- **Versicolor flowers** (green) occupy the middle range with moderate sepal lengths (5.5-7.0 cm)
- **Virginica flowers** (yellow) have the longest sepal lengths (6.3-7.9 cm) and show more variation

### Findings from Histogram (Feature Distributions)
- **Sepal Length**: Roughly bimodal distribution with peaks around 5.0 and 6.5 cm
- **Sepal Width**: More uniformly distributed, concentrated between 2.0-3.5 cm
- **Petal Length**: Highly bimodal with distinct separation between species (1-2 cm for Setosa vs 3-7 cm for others)
- **Petal Width**: Shows clear clustering with three distinct peaks corresponding to each species

### Findings from Box Plots (Feature Ranges by Species)
- **Setosa**: Most compact flower with smallest petals (length: 1-2 cm, width: 0.1-0.6 cm)
- **Versicolor**: Intermediate size with moderate petal dimensions (length: 3-5.5 cm, width: 1-1.8 cm)
- **Virginica**: Largest flower with longest petals (length: 4.5-7 cm, width: 1.4-2.5 cm)
- **Key Observation**: Petal measurements are the **most discriminative features** for species classification
- **Sepal Width**: Shows the least variation between species, making it less useful for classification

### Statistical Insights
- High positive correlation between petal features (length and width) within each species
- Setosa is clearly separable from other species using any feature
- Versicolor and Virginica show some overlap but can be distinguished using petal measurements
- All features show low outliers/outliers, indicating clean, well-measured dataset
